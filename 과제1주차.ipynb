#1주차 과제

  현재 우리는 삶에서 인공지능을 자주 접할 수 있습니다.

 언어, 음성, 이미지, 자율 주행 등에서 인공지능의 발전을 볼 수 있는데, 저는 이 분야에 적용된 기술 및 서비스에 대해 작성할 예정입니다.


*   언어
  * 번역에 인공지능을 적용한 기계번역의 활성화

  솔트룩스라는 기계번역 전문 기업에서는 10가지의 AI 엔진을 제공합니다.
  그 종류로 자연어이해, 음성인식, 음성합성, 대화처리, 심층Q&A, 기계번역, 기계독해, 복합추론, 지식학습, 이미지인식 엔진을 제공합니다.
  
  실제로 사용하고, 기계번역의 정확도를 높이기 위해 학습 과정에 참여한 적이 있습니다. 주로 한국 기사, 외국 기사들을 사용하여 1차로 기계번역으로 검수를 하면, 검수자가 번역의 정확도를 판단하고 문장, 한국인의 문법, 단어에 맞도록 고쳐주는 활동을 하였습니다. 
 4차 산업 활동에 관련해 특강을 해주는 강사를 만난 적이 있는데, 본인은 이제 책을 해외에서 출판할 때 과거와 달리 1차 검수자는 파파고, 솔트룩스, 구글 번역과 같은 기계번역으로 처리를 한 뒤, 2차 검수자에게 오탈자, 문화에 맞는 단어 검수를 맡긴다고 들었습니다.
 
  또한 외국어를 들으면 이어폰에서 원하는 언어로 번역해 주는 시스템 또한 네이버와 같은 기업에서 개발 중에 있습니다. 음성과 언어를 합친 기술들이 하나둘씩 생기고 발전하는 것으로 보아, 분야끼리 결합한 기술과 서비스들이 예상치 못한 곳에서 발전할 것이라 생각합니다.



*   음성
  * TTS(Text-To-Speech)기술을 이용한 음성합성 기술의 등장

  TTS란 삼성에서 발표한 대화형 UI를 추구하는 문자음성 자동 변환 기술입니다. 셀바스에서 발표한 TTS 음성 지능 AI SelvyTTS는 주제와 상황에 맞게 목소리 감정 조절이 가능합니다. 현재 이 기술을 적용한 기계 성우도 존재하는 것으로 보아, 콜센터나 간단한 전화 상담 센터와 같은 경우는 기계로 대체할 수 있을 것이라고 생각합니다. 또한 독거, 비혼 계층이 늘어남에 따라 대화가 필요한 이들에게 목소리의 감정을 인식할 뿐만 아니라, 감정 섞인 목소리로 대화가 가능한 기계의 수요는 높을 것이라 예상합니다.

*   이미지
  * 건강과 이미지

  의료분야에서는 CDSS(Clinical Decision Support System)의 한 방법으로 대량의 이미지와 영상을 분석하는 곳에 인공지능을 사용합니다. 이의 예시는 미국의 FDNA는 구축된 모델을 통해 아이들의 희귀병을 조기 예측하는 서비스와 피부암을 진단하는 것뿐만 아니라 피부암, 해부학, 방사선 전문의를 도와주는 등 다양한 영역에서 찾아 볼 수 있습니다. 주로 의료 영상과 진단 자료들, 환자들을 대상으로 얻은 이미지, 영상들을 토대로 방대한 자료들을 취합하여 결과를 예측하고, 예측결과들의 정확도를 높이는 중입니다. 아마 의료분야에 인공지능 활용도가 높아질수록 오진은 낮아지고 의료비는 더 낮아질 수 있다고 생각합니다. 


*   자율주행
  * 부분 자율 주행 기술 등장
  
 현대와 기아차에서 최초로 부분 자율 주행이 가능한 기술을 내놓았습니다. 이 기술은 자율 주행에 그칠 뿐만 아니라. AI기술이 부합하여 운전자의 운전 성향을 분석하여 차가 스스로 운전자와 흡사한 속도, 스타일로 운전을 할 수 있습니다. 기존의 자율 주행은 운전자가 앞차와의 거리, 운전 패턴을 직접 설정해야 했지만 새로 나온 부분 자율 주행 기술(SCC-ML)의 경우, 자동차가 자동으로 운전자와 거의 동일한 스타일로 운전을 하기 때문에 이질감이나 불편함이 더욱 적을 것이라 생각합니다. 또한 부분 자율주행이다 보니 운전자가 직접 차를 운전하는 경우가 발생하였다면, 자동차는 바로 운전자의 최신 운전 유형을 습득 할 수 있을 것이기 때문에 계속 변화할 수 있습니다. 부분 자율주행을 하는 자동차가 더욱 늘어난다면, 앞 차와의 거리 등을 고려하여 안정성이 더 높은데서 그치지 않고 운전자의 성향까지 분석하기 때문에 실사용자도 늘어날 것이라 예상합니다. 


이 중에서 가장 관심이 가는 분야는 언어 분야입니다. 성경에도 나왔을 정도로 언어의 자유로운 통합은 쉽지 않은 숙제였는데, 인공지능이 발전해감에 따라 해외를 나가서도 휴대폰만 소지하고 있다면 어려움 없이 대화를 할 수 있습니다. 또한, 개인적으로 통번역 계통으로 장래를 희망하였기 때문에 언어와 인공지능을 부합한 기술들에 대해 더욱 자세히 배우는 과정을 목표로 세우고 싶습니다. 앞서 말하였듯, 한 가지 분야의 기술이 아닌, 여러 분야를 접합하거나 아우르는 기술들이 생기고 발전하고 있는데, 다양한 타분야와 언어를 결합한 기술과 서비스를 개발하고 공부해 나가고 싶습니다.
